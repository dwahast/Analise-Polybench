```python
import pandas as pd
df1 = pd.read_csv("idle.csv", index_col=[0])
df1.drop(columns = 'Config', inplace = True)
df2 = pd.read_csv('mali.csv',  index_col=[0])
df1.reset_index(inplace = True)
df2.reset_index(inplace = True)
df1.drop(columns='index', inplace = True)
df2.drop(columns='index', inplace =True)
df2.columns = ['App', 'GPU_Freq', 'Dev', 'Total_Acumulada', 'GPU_Acumulada',
       'A15_Acumulada', 'A7_Acumulada', 'MEM_Acumulada', 'real', 'sys', 'user',
       'EDP', 'A15_Power', 'A7_Power', 'MEM_Power', 'GPU_Power', 'Type']
df2['Host'] = "A15"
df2['A15_Freq'] = '1'
df2['A7_Freq'] = '1'
col = df2['A7_Freq']
df2.drop(labels=['A7_Freq'], axis=1,inplace = True)
df2.insert(0, 'A7_Freq', col)
col = df2['A15_Freq']
df2.drop(labels=['A15_Freq'], axis=1,inplace = True)
df2.insert(0, 'A15_Freq', col)
col = df2['Dev']
df2.drop(labels=['Dev'], axis=1,inplace = True)
df2.insert(0, 'Dev', col)
col = df2['Host']
df2.drop(labels=['Host'], axis=1,inplace = True)
df2.insert(0, 'Host', col)
col = df2['App']
df2.drop(labels=['App'], axis=1,inplace = True)
df2.insert(0, 'App', col)


col = df1['GPU_Acumulada']
df1.drop(labels=['GPU_Acumulada'], axis=1,inplace = True)
df1.insert(0, 'GPU_Acumulada', col)
col = df1['Total_Acumulada']
df1.drop(labels=['Total_Acumulada'], axis=1,inplace = True)
df1.insert(0, 'Total_Acumulada', col)
col = df1['GPU_Freq']
df1.drop(labels=['GPU_Freq'], axis=1,inplace = True)
df1.insert(0, 'GPU_Freq', col)
col = df1['A7_Freq']
df1.drop(labels=['A7_Freq'], axis=1,inplace = True)
df1.insert(0, 'A7_Freq', col)
col = df1['A15_Freq']
df1.drop(labels=['A15_Freq'], axis=1,inplace = True)
df1.insert(0, 'A15_Freq', col)
col = df1['Dev']
df1.drop(labels=['Dev'], axis=1,inplace = True)
df1.insert(0, 'Dev', col)
col = df1['Host']
df1.drop(labels=['Host'], axis=1,inplace = True)
df1.insert(0, 'Host', col)
col = df1['App']
df1.drop(labels=['App'], axis=1,inplace = True)
df1.insert(0, 'App', col)
```
